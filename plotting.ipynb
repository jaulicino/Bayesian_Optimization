```python
import numpy as np

#get some round's data
scores = np.genfromtxt("data/round29_observed_scores.txt")

#by column: VAE, Pyrosetta, DCA, Functionality
```

```python
from matplotlib import pyplot as plt

#plot the VAE probability across rounds
plt.plot(range(len(scores)), (-1)* scores[:,0])
plt.xlabel("Number of Points Sampled (increases with more rounds)")
plt.ylabel("VAE Probability")
plt.title("VAE Score for the Final Scalarization")
```

```python
def is_efficient(costs, multiplier=1):
    costs *= multiplier
    eff_arr = np.ones(costs.shape[0], dtype = bool)

    for i, c in enumerate(costs):
        if eff_arr[i]:
            eff_arr[eff_arr] = np.any(costs[eff_arr]<c, axis=1) 
            eff_arr[i] = True
    costs /= multiplier
    return eff_arr
```

```python
pareto_shell = scores[is_efficient(scores, 1)]
```
